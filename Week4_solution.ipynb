## Data Cleaning and Pre-processing

**Attributes we will be using are: -**

Country code  |	Country name  |  Region  |	 Year  |	population  |	Life expectancy	|  Fertility rate

```python
population.columns
```

```python
years = [str(i) for i in range(1960,2017)]
```

```python
population.drop(['Indicator Name','Indicator Code'],axis = 1, inplace = True)
```

```python
population.columns
```

```python
population.shape
```

```python
population.isnull().sum()
```

```python
population.dropna(axis=0, inplace=True)
```

```python
population.shape
```

```python
population.head()
```

```python
df = pd.melt(population,
             id_vars =['Country Code','Country Name'],
             value_vars = years,
             var_name ='Year',
             value_name = 'Population'
             )
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.tail()
```

```python
df['Country Code'].value_counts()
```

```python
country = country[['Country Code','Region']]
```

```python
country.isnull().sum()
```

```python
country.head()
```

```python
df_merged = pd.merge(country,df,how='left',on='Country Code')
df_merged.head()
```

```python
df_merged = df_merged[['Country Code', 'Country Name', 'Region', 'Year', 'Population']]
```

```python
df_merged.head()
```

```python
df_merged.shape
```

```python
life_expectency.head()
```

```python
life_expectency.columns
```

```python
life_expectency.drop(['Indicator Name','Indicator Code'],axis = 1, inplace = True)
```

```python
life_expectency.columns
```

```python
life_expectency.isnull().sum()
```

```python
life_expectency.dropna(axis=0, inplace=True)
```

```python
life_expectency.shape
```

```python
df_2 = pd.melt(life_expectency,
             id_vars ='Country Code',
             value_vars = years,
             var_name ='Year',
             value_name = 'life_expectency'
             )
```

```python
df_2.head()
```

```python
df_2.shape
```

```python
df_merged.head()
```

```python
df_merged_2 = pd.merge(df_merged,df_2,how='left',on=['Country Code','Year'])
```

```python
df_merged_2.shape
```

```python
fertility.head()
```

```python
fertility.columns
```

```python
fertility.drop(['Indicator Name','Indicator Code'],axis = 1, inplace = True)
```

```python
fertility.columns
```

```python
fertility.isnull().sum()
```

```python
fertility.dropna(axis=0, inplace = True)
```

```python
df_3 = pd.melt(fertility,
             id_vars ='Country Code',
             value_vars = years,
             var_name ='Year',
             value_name = 'fertility'
             )
```

```python
df_merged_3 = pd.merge(df_merged_2,df_3,how='left',on=['Country Code','Year'])
```

## Visualization and Animations

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_merged_3['Year'] = pd.to_numeric(df_merged_3['Year'], errors='coerce')

plt.figure(figsize=(12, 6))
plot = sns.lineplot(x='Year', y='Population', hue='Country Name', data=df_merged_3)
plt.title('Population Over the Years')
plt.xlabel('Year')
plt.ylabel('Population')

# Display x-axis ticks for every 5 years
plot.set_xticks(range(int(df_merged_3['Year'].min()), int(df_merged_3['Year'].max()) + 1, 5))

plt.legend(title='Country Name', bbox_to_anchor=(1.05, 0), loc='lower left', borderaxespad=0.)
plt.show()
```

```python
# Scatter plot for Life Expectancy vs Fertility
plt.figure(figsize=(10, 6))
sns.scatterplot(x='life_expectency', y='fertility', hue='Region', data=df_merged_3)
plt.title('Life Expectancy vs Fertility')
plt.xlabel('Life Expectancy')
plt.ylabel('Fertility')
plt.show()
```

```python
# Bar plot for Average Population by Region
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Population', data=df_merged_3, ci=None)
plt.title('Average Population by Region')
plt.xlabel('Region')
plt.ylabel('Average Population')
plt.show()
```
